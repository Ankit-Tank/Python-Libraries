```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme()
```

```python
# list of all dataset which seaborn has in-built

sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
print(type(tips))
```

```python
tips.head()
```

```python
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    size="size",
    style="smoker"
)
```

```python
#scatter plot using seaborn
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="time"
)
```

```python
#line plot using seaborn
flights=sns.load_dataset("flights")
sns.lineplot(
    data=flights,
    x="year",
    y="passengers"
)
```

```python
#bar plot using seaborn
sns.load_dataset("tips")
sns.barplot(
    data=tips,
    x="day",
    y="tip",
    hue="sex"
)
```

```python
# box plot using seaborn
sns.load_dataset("tips")
sns.boxplot(
    data=tips,
    x="day",
    y="tip"
)
```

```python
#Histogram using seaborn
penguins=sns.load_dataset("penguins")
sns.histplot(
    data=penguins,
    x="body_mass_g",
    bins=30
)
```

```python
#Heatmap - plot rectangular data as color encoded matrix

flights_pivot= flights.pivot(index="month",columns="year",values="passengers")
sns.heatmap(
    data=flights_pivot,
    cmap="coolwarm",
    annot=True,
    fmt="d"
)
plt.title("Passengers Heatmap")
plt.tight_layout()
```
