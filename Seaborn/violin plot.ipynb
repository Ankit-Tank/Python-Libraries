```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print(sns.get_dataset_names())
```

```python
# Load dataset 
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
# Create violin plot

sns.violinplot(x="day", y= "total_bill", data= tips)
plt.title("Total Bill Distribution by Day")
```

```python
sns.violinplot(x="smoker", y= "total_bill", data= tips)
plt.title("Total Bill Distribution by Smoker")
```

```python
sns.violinplot(x="time", y= "total_bill", data= tips)
plt.title("Total Bill Distribution by Time")
```

```python
sns.violinplot(x="sex", y= "time", data= tips)
plt.title("Total Male and Female by time")
```
