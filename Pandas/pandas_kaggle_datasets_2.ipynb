```python
#selecting the data in pandas

#01. selecting the columns
df["city"]
df[["city","temperature","aqi"]]
```

```python
#02.selecting the rows
#a.
df.loc[4876]
df.loc[4876:4881]  #start idx: end idx (inclusive)
```

```python
#b.
df.iloc[5]
df.iloc[0:5]   #start idx: end idx (exclusive)
```

```python
#03.Select rows & columns
df.loc[0, "country"]  # 0th row & specific column
df.iloc[0, 2]         # 1th row & 3rd column (by position)
```

```python
df.iloc[0, 1:5]
df.loc[0, ["country", "aqi"]]

df.loc[0:5, ["country", "city", "latitude"]]   # end is inclusive
df.iloc[0:5, 1:3]                              # end is exclusive
```

```python
#04. Select single scalar value
df.at[0, "city"]
df.iat[0, 2]
```

```python
#Filtering the data
#01. boolean filtering
df[df["aqi"]>100]
```

```python
df[ (df["aqi"]>100) & (df["temperature"]>30)]
```

```python
df[ (df["aqi"]>100) & (df["temperature"]<30)]
```

```python
aqi_data= df[ (df["aqi"]>100) & (df["temperature"]>30)] [["city","aqi"]]
aqi_data
```

```python
#02.Query rules
df.query("aqi>100")
```

```python
df.query("temperature>35")
```

```python
df.query("aqi>100 & temperature >35")[["city","aqi"]]
```
