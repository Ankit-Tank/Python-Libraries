```python
#01.Series:--->
s=pd.Series([1,2,3,4,5])
print(s)
print(type(s))

#index
print(s[0])
print(s[4])
```

```python
s=pd.Series([20,22,24,26] ,index=["Ankit","Sonu","Master","Ram"])
print(s)
print(s["Ankit"])
print(s["Sonu"])
print(s.index)
```

```python
#vectorized operations
s1=pd.Series([1,2,3,4,5])
s2=pd.Series([6,7,8,9,10])
print(s1 + s2)

#mutable value
s1[2]=300
print(s1)

changed_s1=s1.drop(4)     #dropping the value 
print(changed_s1)
```

```python
#02.DataFrame:--->
#creating the DataFrame through dictionary of list
info={
    "Name": ["Ankit","Sonu","Master"],
    "Marks":[300,320,350],
    "Age":[20,22,24]
 }
df=pd.DataFrame(info)
print(df)
print(type(df))

print(df.index)  #labels of rows
print(df.columns)   #to access the columns values 
```

```python
#creating the DataFrame through list of lists
df=pd.DataFrame([["Ankit",20],["Sonu",22],["Master",24]],columns=["Name","Age"])
print(df)
```

```python
#creating the DataFrame through numpy array
import numpy as np
np_arr=np.array([[1,2,3],[4,5,6]])
df=pd.DataFrame(np_arr, columns=["A","B","C"])
print(df)
```
