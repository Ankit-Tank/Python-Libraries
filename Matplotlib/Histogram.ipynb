```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(0)
scores=np.random.normal(165, 7, 500) #normal distribution 

plt.hist(scores)
```

```python
plt.hist(scores, color="red", edgecolor="black")
plt.xlabel("Scores")
plt.ylabel("No. of Student")
plt.title("Scores of Students")
```

```python
plt.hist(scores, color="red", edgecolor="black", bins=20)
plt.xlabel("Scores")
plt.ylabel("No. of Student")
plt.title("Scores of Students")
```

```python
bins = [144, 150, 156, 162, 168, 174, 180, 186]
plt.hist(scores, color="red", edgecolor="black", bins=bins)
plt.xlabel("Scores")
plt.ylabel("No. of Student")
plt.title("Scores of Students")
```

```python
# Multiple datasets
legit_transactions = [
2.99, 5.49, 8.99, 12.50, 14.99, 19.99, 23.45, 29.99, 34.99, 39.50,
45.00, 49.99, 55.25, 60.00, 75.99, 89.99, 120.50, 150.00, 199.99,
249.99, 300.75, 450.00, 600.00, 850.00, 1200.00
]

fraud_transactions = [
50, 100, 150, 200, 300, 500, 500, 750, 1000, 1000,
1200, 1500, 1500, 2000, 2500, 3000, 3000
]
plt.hist(fraud_transactions, color="red",bins=20, label="Fraud Transtions", edgecolor="black", alpha= 0.6)
plt.hist(legit_transactions, color="green",bins=20, label="Legit Transtions", edgecolor="black", alpha= 0.6)
plt.title("Legit vs Fraud Transactions")
plt.xlabel("Amount (in $)")
plt.ylabel("Frequency")
plt.grid()
plt.legend()
```
