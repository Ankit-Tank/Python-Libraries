```python
# Real data

# Marvel Movies (Oscar Winners) 

oscar_marvel_movies = [
    "Spider-Man 2",
    "Big Hero 6",
    "Black Panther",
    "Black Panther: Wakanda Forever",
    "Guardians of the Galaxy Vol. 3"
]

years = [2004, 2014, 2018, 2022, 2023]

oscar_marvel_revenue = [789, 657, 1347, 859, 845]  # in $M


plt.plot(years,oscar_marvel_revenue)

plt.title("Marvel Movies Revenue Each Years")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
```

```python
# Non-Oscar Winner Marvel Movies
non_oscar_marvel_movies = [
    "Iron Man 2",
    "Thor",
    "Captain America: The First Avenger",
    "Doctor Strange",
    "Spider-Man: Homecoming"
]

non_oscar_marvel_revenue = [623, 449, 370, 677, 880]  # in $M

plt.plot(years,oscar_marvel_revenue,"o--r",linewidth=2, label="Oscar Movies")
plt.plot(years,non_oscar_marvel_revenue,"^--g",label="Non-Oscar Movies")

plt.title("Marvel Movies Revenue Each Years")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")

plt.legend()
```
