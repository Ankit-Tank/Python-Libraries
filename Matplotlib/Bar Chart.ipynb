```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Bar Chart

Marvel_Movies = [
    "Spider-Man 2",
    "Big Hero 6",
    "Black Panther",
    "Black Panther: Wakanda Forever",
    "Guardians of the Galaxy Vol. 3"
]

Years = [2012, 2014, 2016, 2018, 2020]
Marvel_Revenue = [789, 657, 1347, 859, 845]  # in $M

plt.bar(Years,Marvel_Revenue,color="red")

plt.title("Marvel Movies Revenue Each Years")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")

plt.tight_layout()
```

```python
plt.figure()

plt.bar(Years, Marvel_Revenue, color="blue")

plt.title("Marvel Movies Revenue Each Years")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")

for i in range(len(Years)):
    plt.text(Years[i], Marvel_Revenue[i]+50, str(Marvel_Revenue[i]), ha="center")

plt.ylim(0,1600)
plt.xlim(2010,2022)
```

```python
oscar_marvel_movies = [
    "Spider-Man 2",
    "Big Hero 6",
    "Black Panther",
    "Black Panther: Wakanda Forever",
    "Guardians of the Galaxy Vol. 3"
]

years = [2004, 2014, 2018, 2022, 2023]

oscar_marvel_revenue = [789, 657, 1347, 859, 845]  # in $M

non_oscar_marvel_movies = [
    "Iron Man 2",
    "Thor",
    "Captain America: The First Avenger",
    "Doctor Strange",
    "Spider-Man: Homecoming"
]

non_oscar_marvel_revenue = [623, 449, 370, 677, 880]  # in $M

x=np.arange(len(years))
width=0.3

plt.bar(x-width/2,oscar_marvel_revenue,width, color="blue",label="Oscar Movies")
plt.bar(x+width/2,non_oscar_marvel_revenue,width, color="red",label="Non Oscar Movies")

plt.title("Marvel Movies Revenue Each Years")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.legend()
plt.xticks(x, years)
```
