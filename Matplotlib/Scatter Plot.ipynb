```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
#scatter plot
people = ["Person A", "Person B", "Person C", "Person D", "Person E",
          "Person F", "Person G", "Person H", "Person I", "Person J"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
bp = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]

plt.scatter (age, bp)
plt.title("Age v/s Blood Pressure")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
```

```python
plt.scatter (age, bp, s=bp ,marker ="^", color="green", alpha =0.5)
plt.title("Age v/s Blood Pressure")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.grid()
```

```python
colors=["green" if x <130 else "red" for x in bp]
plt.scatter (age, bp, s=bp , color=colors)
plt.title("Age v/s Blood Pressure")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.grid()
```

```python
plt.scatter (age, bp, s=bp ,cmap="YlOrRd", c=bp)
plt.title("Age v/s Blood Pressure")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.grid()
plt.colorbar(label="BP")
```

```python
plt.scatter (age, bp, s=bp ,cmap="YlOrRd", c=bp)
plt.title("Age v/s Blood Pressure")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.grid()
plt.colorbar(label="BP")

for i in range(len(people)):
    plt.annotate(people[i], xy=(age[i],bp[i]), xytext=(age[i]+1,bp[i]+1))

plt.xlim(min(age),max(age)+10)
plt.ylim(min(bp),max(bp)+5)
```
