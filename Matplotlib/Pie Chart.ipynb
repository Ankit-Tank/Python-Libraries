```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
#Pie Chart

expenses = ["Salaries", "Rent", "Marketing", "R&D", "Miscellaneous"]
amounts = [500, 150, 120, 100, 50]

explode=[0,0,0.1,0,0.3]

plt.pie(amounts, labels=expenses , autopct="%1.1f%%",
       wedgeprops={
           "edgecolor": "black",
           "linewidth": 2
       },
        explode=explode,
        startangle=90,
        shadow=True
       )
```
